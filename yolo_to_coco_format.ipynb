#### coco format bbox coordinates xmin ymin width height

```python
from glob import glob
import os
import numpy as np
import json

type = 'train_with_background_images'
# directory = f'/home/shataxi.dubey/shataxi_work/vlm_on_planet/lucknow_train_test_split/{type}'
# # images_path = sorted(glob(f'{directory}/images/*'))
# images_path = sorted(glob(f'/home/shataxi.dubey/shataxi_work/vlm_on_planet/Florence-2/dynamic_lucknow_coco_train_test/{type}/*'))
# labels_path = f'{directory}/labels'

directory = f'/home/shataxi.dubey/shataxi_work/vlm_on_planet/gms/west_bengal_small_639_sq_km/kiln_images/{type}'
images_path = sorted(glob(f'{directory}/images/*'))
labels_path = f'{directory}/labels'

# directory = f'/home/shataxi.dubey/shataxi_work/vlm_on_planet/gms/west_bengal_small_639_sq_km/kiln_images/{type}'
# images_path = sorted(glob(f'{directory}/images/*'))
# labels_path = f'{directory}/labels'
image_size = 320
```

```python
coco_dict = {}

coco_dict['info'] = {
    "year": "2024",
    "version": "0",
    "description": f"coco format of yolo labels of {type} set",
    "contributor": "",
    "url": "",
    "date_created": ""
  }

coco_dict['licenses'] = [{
      "id": 1,
      "url": "",
      "name": ""
    }]

coco_dict['categories'] = [
    {
      "id": 0,
      "name": "brick kilns with chimney",
      "supercategory": "none"
    },
    # {
    #   "id": 1,
    #   "name": "background",
    #   "supercategory": "none"
    # },
]

coco_dict['images'] = []

coco_dict['annotations'] = []
```

```python
coco_dict
```

```python
def add_bboxes(label_path, image_id, annotation_id):
    bbox_dict = []

    bboxes = np.loadtxt(label_path, ndmin = 2)
    bboxes = bboxes * image_size

    for bbox in bboxes:
        xmin = min(bbox[[1,3,5,7]])
        ymin = min(bbox[[2,4,6,8]])
        xmax = max(bbox[[1,3,5,7]])
        ymax = max(bbox[[2,4,6,8]])
        width = int(xmax - xmin)
        height = int(ymax - ymin)
        bbox_dict.append({"id": annotation_id,
        "image_id": image_id,
        "category_id": 0,
        "bbox": [
            int(xmin),
            int(ymin),
            width,
            height
            ],
        "area": width * height,
        "segmentation": [],
        "iscrowd": 0
        })
        annotation_id += 1
    
    return bbox_dict, annotation_id
```

```python
annotation_id = 0
for image_id, image_path in enumerate(images_path):
    image_name = os.path.basename(image_path)
    if image_name[-3:] == 'tif' or image_name[-3:] == 'png':
      coco_dict['images'].append({
        "id": image_id,
        "license": 1,
        "file_name": image_name,
        "height": image_size,
        "width": image_size,
        "date_captured": ""
      })

      label_path = os.path.join(labels_path, image_name[:-4]+'.txt')
      if os.path.exists(label_path):
        bbox_dict, annotation_id = add_bboxes(label_path, image_id, annotation_id)
        coco_dict['annotations'] += bbox_dict
      # else:
      #   coco_dict['annotations'].append({"id": annotation_id,
      #                               "image_id": image_id,
      #                               "category_id": 1,
      #                               "bbox": [0, 0, image_size, image_size], # earlier with planet imagery, I was using this annotation
      #                               "area": image_size * image_size,
      #                               "segmentation": [],
      #                               "iscrowd": 0
      #                               })
      # but now using this annotation as per https://blog.roboflow.com/missing-and-null-image-annotations/
      # no annotations if there is no object in the image
      
        annotation_id += 1
```
